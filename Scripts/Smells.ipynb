```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
df = pd.read_csv('Smells.csv')
sns.heatmap(df.drop(columns='Things'), yticklabels=df['Things'])
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
df2 = df.corr()
print(df2.columns)
```

```python
corr, p_values = scipy.stats.spearmanr(df['Alan'], df['Drew'])
print(corr, p_values)
```

```python
Cspearman = pd.DataFrame()
Pspearman = pd.DataFrame()
for c in df2.columns:
    corrlist = []
    pvals = []
    for cc in df2.columns:
        corr, p_values = scipy.stats.spearmanr(df[c], df[cc])
        corrlist.append(corr)
        pvals.append(p_values)
    Cspearman[c] = corrlist
    Pspearman[c] = pvals
```

```python
print(Cspearman.head())
print(Pspearman.head())
```

```python
sns.heatmap(Cspearman,yticklabels=df2.columns,annot=True,cmap="GnBu").set_title('Monoticity of Data')
```

```python
sns.heatmap(Pspearman,yticklabels=df2.columns,annot=True,cmap="GnBu").set_title('Pvalues (monoticity)')
```

```python
Cpearson = pd.DataFrame()
Ppearson = pd.DataFrame()
for c in df2.columns:
    corrlist = []
    pvals = []
    for cc in df2.columns:
        corr, p_values = scipy.stats.pearsonr(df[c], df[cc])
        corrlist.append(corr)
        pvals.append(p_values)
    Cpearson[c] = corrlist
    Ppearson[c] = pvals
print(Cpearson.head())
print(Ppearson.head())
```

```python
sns.heatmap(Cpearson,yticklabels=df2.columns,annot=True,cmap="GnBu").set_title('PCC of Data')
```

```python
sns.heatmap(Ppearson,yticklabels=df2.columns,annot=True,cmap="GnBu").set_title('PCC Pvals')
```
