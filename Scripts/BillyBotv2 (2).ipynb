```python
w = 4

def generateTable(data,k):

    T = {}
    for i in range(len(data)-k):
        X = data[i:i+k]
        Y = data[i+k]
        #print("X  %s and Y %s  "%(X,Y))

        #print("First", T.get(X))
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        else:
            #print("Second", T[X].get(Y))
            if T[X].get(Y) is None:
                T[X][Y] = 1
            else:
                T[X][Y] += 1

    return T

T = generateTable("bingus bigchungus bugs", k=w)
print(T)
```

```python
def convertFreqIntoProb(T):
    for kx in T.keys():
        s = float(sum(T[kx].values()))
        #print(T.keys())
        #print(T[kx].keys())
        for k in T[kx].keys():
            T[kx][k] = T[kx][k]/s

    return T

T = convertFreqIntoProb(T)
print(T)
```

```python
# text_path = "/content/drive/My Drive/barstoo.txt"
text_path = 'barstoo.txt'
def load_text(filename):
    with open(filename,encoding='utf8') as f:
        return f.read().lower()

text = load_text(text_path)
print(text)
```

```python
def MarkovChain(text,k):
    T = generateTable(text,k)
    T = convertFreqIntoProb(T)
    return T

model = MarkovChain(text, k=w)
print(model)
```

```python
import numpy as np

def sample_next(ctx,model,k):

    ctx = ctx[-k:]
    if model.get(ctx) is None:
        return " "
    possible_Chars = list(model[ctx].keys())
    possible_values = list(model[ctx].values())

    #print(possible_Chars)
    #print(possible_values)

    return np.random.choice(possible_Chars,p=possible_values)

sample_next("banana",model,k=w)
```

```python
def generateText(starting_sent,k,maxLen=1000):

    sentence = starting_sent
    ctx = starting_sent[-k:]

    for ix in range(maxLen):
        next_prediction = sample_next(ctx,model,k)
        sentence += next_prediction
        ctx = sentence[-k:]
    return sentence

print("Function Created Successfully!")

text2 = generateText("lickspittle bitc",k=w,maxLen=500)
print(text2)
```
